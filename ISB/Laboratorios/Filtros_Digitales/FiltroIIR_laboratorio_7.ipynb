```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display
import sympy
```

### Dominio de la frecuencia de las señales

```python
N = 2**10  #10 bits
F_obs2 = 0.001*len(d_obs2[15])  #frecuencia de la data, ya que la frecuencia con la que se obtuvo los datos en
#el Bitalino es 1000Hz y cada onda PQRST tiene una freq de la longitud de cada subarray
X_d_obs2 = np.fft.fft(d_obs2[15],N) #FFT de la señal
X_d_obs2 = X_d_obs2[0:N//2]
X_d_obs2m = np.abs(X_d_obs2) #valor absoluto de la FFT
F_obs2 = np.linspace(0,10*F_obs2/2,N//2)

plt.plot(F_obs2, X_d_obs2m)
plt.grid(linestyle=":")
plt.title("Analisis espectral de las señales")
plt.xlabel("Frecuencias (hz)")
plt.ylabel("FFT (db)")
plt.ylim(0,10000)
plt.xlim(0,5)

peaks2, properties2 = signal.find_peaks(X_d_obs2m, height=5000)
plt.plot(F_obs2[peaks2],properties2["peak_heights"],"x")
plt.text(F_obs2[peaks2][0],properties2["peak_heights"][0], f"{np.round(F_obs2[peaks2][0],2)} hz")
plt.text(F_obs2[peaks2][1],properties2["peak_heights"][1], f"{np.round(F_obs2[peaks2][1],2)} hz")
plt.show()

#se tiene un valor muy alto cercano a 0Hz, esto posiblemente se deba a una corriente DC que tiene
#valores muy altos a una frecuencia 0.
```

### Diseño del filtro pasa bajo

```python
wp= 94
ws= 157 #estos dos según indicado en la guia

N_ord, Wc= signal.buttord(wp, ws, gpass=3, gstop=40, analog=True) #gpass=loss in passband edge frecuencies, gstop= attenuation in stopband edge frecuencies 
#Para encontrar el mejor orden y la mejor frecuencia angular de corte para el filtro con estos parametros
fc = np.round(Wc/(2*np.pi) ,2) 

print(f"Orden del filtro: {N_ord}")
print(f"Frecuencia de corte calculada: {fc} hz")
print(f"Wc: {Wc} hz")
```

```python
# Funcion de transferencia del filtro pasa bajo analogico
b,a = signal.butter(4, Wc, 'lowpass', True, 'ba') #usamos un orden 4
#porque es mejor la respuesta al momento del filtrado (prueba y error)

#np.set_printoptions(precision=3)
b,a
```

```python
# Respuesta en frecuencia del filtro analogico
Fmax=20 #un límite para la grafica
F_freqs = np.linspace(0,Fmax,1000)
W_freqs = 2*np.pi*F_freqs  #Angular frecuences

#creamos plot para la respuesta del filtro a frecuencias y se observa que la frecuencia de corte es la deseada
W_freqs, H = signal.freqs(b,a,W_freqs)
Hm=np.abs(H)
plt.plot(F_freqs, 20 * np.log10(Hm))
plt.grid(linestyle=":")
plt.xlabel("Frecuencia (hz)")
plt.ylabel("Magnitud (db)")
```

### Transformada bilineal de H(s) a H(z)

```python
bd, ad = signal.bilinear(b,a,Fs)

Wz, Hd = signal.freqz(bd, ad, W_freqs/Fs)

Hdm = np.abs(Hd)
plt.plot(F_freqs, 20 * np.log10(Hdm))
plt.grid(linestyle=":")
plt.xlabel("Frecuencia (hz)")
plt.ylabel("Magnitud (db)")
```

### Filtrado
